### Imputing the missing values

```python
# Caterozing categorical anf Numerical Feature
cat_cols = dataframe.columns[dataframe.nunique() < 5]
num_cols = dataframe.columns[dataframe.nunique() >= 5]
cat_cols= dataframe[cat_cols]
num_cols= dataframe[num_cols]

# imputation in Categorical Features
cat_cols= cat_cols.fillna(cat_cols.mode().iloc[0])

from sklearn.impute import SimpleImputer
categorical_imputer = SimpleImputer(strategy='most_frequent')
categorical_imputer.fit(cat_cols)
```

```python
cal_cols = pd.DataFrame(categorical_imputer.transform(cat_cols), columns=cat_cols.columns)
cat_cols
```

#### Numerical Feature( Imputation)

```python
num_cols = num_cols.fillna(num_cols.mean())
num_imputer = SimpleImputer(strategy='median')
num_imputer.fit(num_cols[num_cols.columns])
```

```python
num_cols = pd.DataFrame ( num_imputer.transform(num_cols), columns= num_cols.columns)
num_cols
```

```python
print(cat_cols.isna().sum())
print(num_cols.isna().sum())
```
